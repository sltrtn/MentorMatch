```python
import csv
import random

# Indian names (matching the regional diversity in your dataset)
first_names = [
    # Male names
    "Rajesh", "Amit", "Suresh", "Vikram", "Arun", "Karthik", "Ravi", "Sandeep", 
    "Manoj", "Deepak", "Anand", "Vijay", "Sanjay", "Rahul", "Nikhil", "Rohan",
    "Arjun", "Varun", "Ashok", "Pradeep", "Ramesh", "Krishna", "Siddharth", "Aditya",
    "Naveen", "Harish", "Prakash", "Ganesh", "Mohan", "Sunil", "Anuj", "Vishal",
    # Female names
    "Priya", "Sneha", "Anjali", "Kavya", "Divya", "Meera", "Swathi", "Lakshmi",
    "Pooja", "Asha", "Rekha", "Suma", "Nisha", "Radha", "Deepa", "Sangeetha",
    "Anita", "Seema", "Veena", "Maya", "Preeti", "Shruti", "Neha", "Ritu",
    "Shalini", "Vandana", "Archana", "Jyoti", "Smita", "Bharti", "Sarita", "Pallavi"
]

last_names = [
    "Kumar", "Sharma", "Reddy", "Patel", "Singh", "Iyer", "Nair", "Das",
    "Gupta", "Verma", "Rao", "Menon", "Joshi", "Pillai", "Mehta", "Shah",
    "Desai", "Agarwal", "Banerjee", "Chatterjee", "Krishnan", "Naik", "Kulkarni", "Bhat",
    "Chopra", "Malhotra", "Kapoor", "Srinivasan", "Varma", "Murthy", "Dutta", "Sen",
    "Bose", "Pandey", "Mishra", "Saxena", "Trivedi", "Thakur", "Chauhan", "Rajan",
    "Mohan", "Venkatesh", "Ramachandran", "Subramanian", "Anand", "Prasad", "Hegde", "Shetty"
]

def generate_unique_names(count):
    """Generate unique full names"""
    used_names = set()
    names = []
    
    while len(names) < count:
        first = random.choice(first_names)
        last = random.choice(last_names)
        full_name = f"{first} {last}"
        
        if full_name not in used_names:
            used_names.add(full_name)
            names.append(full_name)
    
    return names

# Read the CSV file
input_file = 'mentors_dataset.csv'  # Update with your file path
output_file = 'mentors_updated.csv'

# Generate 500 unique names
unique_names = generate_unique_names(500)

# Read and update the CSV
with open(input_file, 'r', encoding='utf-8') as infile, \
     open(output_file, 'w', encoding='utf-8', newline='') as outfile:
    
    reader = csv.DictReader(infile)
    fieldnames = reader.fieldnames
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)
    
    writer.writeheader()
    
    for i, row in enumerate(reader):
        # Update the name field
        row['name'] = unique_names[i]
        writer.writerow(row)

print(f"✓ Successfully updated {len(unique_names)} mentor names!")
print(f"✓ Output saved to: {output_file}")
print(f"\nSample names generated:")
for i in range(10):
    print(f"  Mentor {i+1}: {unique_names[i]}")
```
